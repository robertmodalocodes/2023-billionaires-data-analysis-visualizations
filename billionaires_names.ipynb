```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
from matplotlib.patches import Patch
import geopandas as gpd

%matplotlib inline

pd.set_option("display.max_columns", None)
sns.set_style('darkgrid')
```

```python
df_bill = pd.read_csv("./Billionaires Statistics Dataset.csv")
```

```python
df_bill.head()
```

```python
# Billionaires names
df_bill_names = df_bill.sort_values(by='finalWorth', ascending=False).copy()
```

```python
df_bill_names.head()
```

```python
df_bill_names.dtypes
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=df_bill_names.head(20), x='finalWorth', y='personName', hue='category', palette='tab10')
plt.title("Top 20 billionaires by industry", pad=20)
#plt.xticks([i for i in range(0, 12, 1)], [i for i in month_names.values()], rotation=0)
plt.ylabel('Billionaires', labelpad=15)
plt.xlabel('Final Worth (in million USD)', labelpad=15)
plt.legend(title="Industry")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=df_bill_names.head(20), x='finalWorth', y='personName', hue='country', palette='tab10')
plt.title("Top 20 billionaires by country", pad=20)
#plt.xticks([i for i in range(0, 12, 1)], [i for i in month_names.values()], rotation=0)
plt.ylabel('Billionaires', labelpad=15)
plt.xlabel('Final Worth (in million USD)', labelpad=15)
plt.legend(title="Country")
plt.show()
```
